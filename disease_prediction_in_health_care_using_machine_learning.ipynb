# Task
Generate two synthetic datasets, "diabetes_dataset_1.csv" and "diabetes_dataset_2.csv," each with 300 samples for diabetic patients including features like Glucose, BloodPressure, BMI, Age, and an Outcome. Then, load these datasets, train a classification model on "diabetes_dataset_1.csv," and evaluate its performance on both datasets. Finally, compare the model's performance on the two datasets using visualizations and print classification reports and confusion matrices for both evaluations.

## Generate Dataset 1

### Subtask:
Generate a synthetic dataset of 300 samples for diabetic patients, including features like Glucose, BloodPressure, BMI, Age, and an Outcome (0 for non-diabetic, 1 for diabetic).


**Reasoning**:
The first step is to import the necessary libraries, `make_classification` for synthetic data generation and `pandas` for DataFrame manipulation, as specified in the instructions.



```python
from sklearn.datasets import make_classification
import pandas as pd

print("Libraries imported successfully.")
```

**Reasoning**:
Now that the libraries are imported, the next step is to generate the synthetic dataset with 300 samples, 4 informative features, and 2 classes, then convert it into a pandas DataFrame with the specified column names and an 'Outcome' column.



```python
X, y = make_classification(n_samples=300, n_features=4, n_informative=4, n_redundant=0, n_repeated=0, n_classes=2, random_state=42)
df_diabetes_1 = pd.DataFrame(X, columns=['Glucose', 'BloodPressure', 'BMI', 'Age'])
df_diabetes_1['Outcome'] = y

print("Synthetic dataset 'df_diabetes_1' generated successfully with 300 samples.")
print(df_diabetes_1.head())
```

## Save Dataset 1 to CSV

### Subtask:
Save the first generated dataset to a CSV file named 'diabetes_dataset_1.csv'.


**Reasoning**:
Save the `df_diabetes_1` DataFrame to a CSV file named 'diabetes_dataset_1.csv', ensuring the index is not included as specified.



```python
df_diabetes_1.to_csv('diabetes_dataset_1.csv', index=False)
print("Dataset 'df_diabetes_1' saved to 'diabetes_dataset_1.csv' successfully.")
```

## Generate Dataset 2

### Subtask:
Generate a second synthetic dataset of 300 samples for diabetic patients with similar features as Dataset 1.


**Reasoning**:
Generate a second synthetic dataset, `df_diabetes_2`, with 300 samples, 4 informative features, and 2 classes, using a different random_state (43) to ensure it's distinct from the first dataset. Then, convert the features into a pandas DataFrame with specified column names and add the 'Outcome' column. Finally, print a confirmation message and display the head of the new DataFrame.



```python
X2, y2 = make_classification(n_samples=300, n_features=4, n_informative=4, n_redundant=0, n_repeated=0, n_classes=2, random_state=43)
df_diabetes_2 = pd.DataFrame(X2, columns=['Glucose', 'BloodPressure', 'BMI', 'Age'])
df_diabetes_2['Outcome'] = y2

print("Synthetic dataset 'df_diabetes_2' generated successfully with 300 samples.")
print(df_diabetes_2.head())
```

## Save Dataset 2 to CSV

### Subtask:
Save the second generated dataset to a CSV file named 'diabetes_dataset_2.csv'.


**Reasoning**:
Save the `df_diabetes_2` DataFrame to a CSV file named 'diabetes_dataset_2.csv', ensuring the index is not included as specified.



```python
df_diabetes_2.to_csv('diabetes_dataset_2.csv', index=False)
print("Dataset 'df_diabetes_2' saved to 'diabetes_dataset_2.csv' successfully.")
```

## Load Datasets

### Subtask:
Load both 'diabetes_dataset_1.csv' and 'diabetes_dataset_2.csv' into separate pandas DataFrames.


**Reasoning**:
Load the 'diabetes_dataset_1.csv' file into a pandas DataFrame named `df_diabetes_1_loaded`.



```python
df_diabetes_1_loaded = pd.read_csv('diabetes_dataset_1.csv')
print("Dataset 'diabetes_dataset_1.csv' loaded successfully.")
```

**Reasoning**:
Load the 'diabetes_dataset_2.csv' file into a pandas DataFrame named `df_diabetes_2_loaded`, and then print the first 5 rows of both `df_diabetes_1_loaded` and `df_diabetes_2_loaded` to verify successful loading as per the instructions.



```python
df_diabetes_2_loaded = pd.read_csv('diabetes_dataset_2.csv')
print("Dataset 'diabetes_dataset_2.csv' loaded successfully.\n")

print("First 5 rows of df_diabetes_1_loaded:")
print(df_diabetes_1_loaded.head())
print("\nFirst 5 rows of df_diabetes_2_loaded:")
print(df_diabetes_2_loaded.head())
```

## Prepare Data and Train-Test Split (Dataset 1)

### Subtask:
Split Dataset 1 into features (X1) and target (y1). Then, split X1 and y1 into training and testing sets (X1_train, X1_test, y1_train, y1_test).


**Reasoning**:
To prepare the data for model training, I need to separate the features from the target variable and then split them into training and testing sets. First, I'll import the `train_test_split` function.



```python
from sklearn.model_selection import train_test_split

print("train_test_split imported successfully.")
```

**Reasoning**:
Now that `train_test_split` is imported, I will separate the features and target from `df_diabetes_1_loaded` and then split them into training and testing sets as per the subtask instructions.



```python
X1 = df_diabetes_1_loaded[['Glucose', 'BloodPressure', 'BMI', 'Age']]
y1 = df_diabetes_1_loaded['Outcome']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=42)

print("Dataset 1 successfully split into training and testing sets.")
print(f"X1_train shape: {X1_train.shape}")
print(f"X1_test shape: {X1_test.shape}")
print(f"y1_train shape: {y1_train.shape}")
print(f"y1_test shape: {y1_test.shape}")
```

## Prepare Data and Train-Test Split (Dataset 2)

### Subtask:
Split Dataset 2 into features (X2) and target (y2). Then, split X2 and y2 into training and testing sets (X2_train, X2_test, y2_train, y2_test).


**Reasoning**:
As instructed, I will now separate the features and target from `df_diabetes_2_loaded` and then split them into training and testing sets using `train_test_split` with the specified parameters.



```python
X2 = df_diabetes_2_loaded[['Glucose', 'BloodPressure', 'BMI', 'Age']]
y2 = df_diabetes_2_loaded['Outcome']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=42)

print("Dataset 2 successfully split into training and testing sets.")
print(f"X2_train shape: {X2_train.shape}")
print(f"X2_test shape: {X2_test.shape}")
print(f"y2_train shape: {y2_train.shape}")
print(f"y2_test shape: {y2_test.shape}")
```

## Define and Train Model

### Subtask:
Define a classification model (e.g., Logistic Regression or RandomForestClassifier) and train it using X1_train and y1_train.


**Reasoning**:
To define and train a classification model, I first need to import the `LogisticRegression` class from `sklearn.linear_model`.



```python
from sklearn.linear_model import LogisticRegression

print("LogisticRegression imported successfully.")
```

**Reasoning**:
Now that `LogisticRegression` is imported, I will instantiate the model with `random_state=42` and then train it using `X1_train` and `y1_train` as instructed.



```python
model = LogisticRegression(random_state=42, solver='liblinear')
model.fit(X1_train, y1_train)

print("Logistic Regression model trained successfully on Dataset 1.")
```

## Evaluate Model on Dataset 1

### Subtask:
Evaluate the trained model's performance (e.g., accuracy, precision, recall) on X1_test and y1_test. Print the classification report and confusion matrix.


**Reasoning**:
To evaluate the model and visualize the confusion matrix, I need to import `classification_report`, `confusion_matrix` from `sklearn.metrics`, `matplotlib.pyplot` as `plt`, and `seaborn` as `sns`.



```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("Required libraries for evaluation and visualization imported successfully.")
```

**Reasoning**:
Now that the libraries are imported, I will use the trained model to make predictions on `X1_test`, print the classification report, and then generate and display the confusion matrix using a seaborn heatmap for better visualization, as per the instructions.



```python
y1_pred = model.predict(X1_test)

print("Classification Report for Dataset 1:")
print(classification_report(y1_test, y1_pred))

cm_d1 = confusion_matrix(y1_test, y1_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_d1, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix - Dataset 1')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

## Evaluate Model on Dataset 2

### Subtask:
Evaluate the *same* trained model's performance on X2_test and y2_test. Print the classification report and confusion matrix to compare with Dataset 1.


**Reasoning**:
Now, I will use the previously trained model to make predictions on the `X2_test` data, generate the classification report, and visualize the confusion matrix for Dataset 2, following the instructions.



```python
y2_pred = model.predict(X2_test)

print("Classification Report for Dataset 2:")
print(classification_report(y2_test, y2_pred))

cm_d2 = confusion_matrix(y2_test, y2_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_d2, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix - Dataset 2')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

## Visualize Model Performance Comparison

### Subtask:
Create a bar chart to visualize and compare the accuracy or other relevant metrics of the model on both Dataset 1 and Dataset 2. Include appropriate legends.


**Reasoning**:
I need to create a bar chart to visualize the accuracy scores for Dataset 1 and Dataset 2. I will prepare the data as a pandas Series and then use matplotlib to generate the bar chart with appropriate labels, title, and text annotations.



```python
accuracies = pd.Series({
    'Dataset 1': 0.84,
    'Dataset 2': 0.39
})

plt.figure(figsize=(7, 5))
ax = accuracies.plot(kind='bar', color=['skyblue', 'lightcoral'])

plt.title('Model Accuracy Comparison: Dataset 1 vs. Dataset 2')
plt.xlabel('Dataset')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.xticks(rotation=0)

for i, v in enumerate(accuracies):
    ax.text(i, v + 0.02, f'{v:.2f}', color='black', ha='center')

plt.tight_layout()
plt.show()

print("Bar chart visualizing model accuracy comparison displayed.")
```
